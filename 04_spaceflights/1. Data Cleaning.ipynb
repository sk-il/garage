```python
def categorize_orbits(df):
    # Define orbit groups
    orbit_groups = {
        'Low Earth Orbits': ['Low Earth', 'LEO', 'SSO'],
        'Medium and High Earth Orbits': ['Medium Earth', 'High Earth', 'MEO', 'HEO'],
        'Geosynchronous and Geostationary Orbits': ['Geostationary', 'GEO', 'GSO', 'IGSO', 'Sub-GSO', 'GTO'],
        'Highly Elliptical Orbits': ['Highly elliptical', 'Extremely Elliptical', 'Extremely elliptical', 
                                     'Elliptical High Earth'],
        'Molniya and Tundra Orbits': ['Molniya', 'Tundra', 'Tundra/Quasi-Zenith'],
        'Miscellaneous Orbits': ['Lunar free-return', 'Salyut 6', 'Transatmospheric', 'Transatmospheric Earth', 
                                 'Galactocentric', 'Intelsat', 'Low Lunar', 'Sun–Earth L1', 'Sun–Earth L2', 'Earth–Moon L2',
                                 'Lunar', 'Areocentric', 'Cytherocentric', 'Mercurian', 'Jovicentric', 'Cytherean', 
                                 'Jovian', 'TLI', 'TMI', 'Lunar transfer', 'Moon transfer', 'TMI to lunar surface', 
                                 'TMI to Martian surface', 'TMI to Areocentric', 'TLI to lunar surface', 
                                 'TMI to Martian Surface'],
        'Suborbital': ['Suborbital']
    }

    # Apply the mappings
    for group, orbits in orbit_groups.items():
        df.loc[df['payload_orbit'].isin(orbits), 'payload_orbit'] = group

    return df['payload_orbit']


# Categorize the payload orbits
df['payload_orbit'] = categorize_orbits(df)

df.groupby('payload_orbit')['flight_n'].nunique().sort_values(ascending=False).reset_index()
```

```python
def categorize_payload_functions(df, mapping):
    # Standardize the 'payload_function' to lowercase for uniformity
    df['payload_function'] = df['payload_function'].str.lower().fillna('Unknown')

    # Apply the mappings
    for category, keywords in mapping.items():
        pattern = '|'.join(keywords)  # Create a regex pattern to match any of the keywords
        df.loc[df['payload_function'].str.contains(pattern), 'payload_function'] = category

    # Assign all uncategorized functions to 'Other'
    categories = mapping.keys()
    df.loc[~df['payload_function'].isin(categories), 'payload_function'] = 'Other'

    return df['payload_function']


# Define the mapping dictionary based on our discussion
mapping = {
    'Missile and Defense': ['missile', 'abm', 'rv test', 'attack', 'target', 'interceptor'],
    'Communications': ['communication', 'amateur radio'],
    'Researches and Observations': ['research', 'observation', 'aeronomy', 'ionospheric', 'magnetospheric', 'plasma', 
                                    'ionosphere', 'solar', 'radiation', 'astronomy', 'imaging', 'remote sensing', 
                                    'education',  'microgravity', 'auroral', 'atmospheric', 'biological', 'oceanography', 
                                    'geospace', 'heliophysics', 'atmospheric science', 'air density', 'fields',
                                    'lunar', 'mars', 'venus', 'scientific', 'experiment', 'film return', 'imagery',
                                    'telescope'],
    'Technology': ['technology', 'calibration', 'iot', 'cubesat deployer'],
    'Meteorology': ['meteorology', 'weather', 'early warning', 'climatology'],
    'Tests': ['test', 'boilerplate'],
    'Intelligence and Surveillance': ['reconnaissance', 'elint', 'sigint'],
    'Navigation and Positioning': ['navigation', 'geodesy', 'ais'],
    'Logistics and Operations': ['logistic', 'iss ', ' iss','satellite deployment', 'space station', 'crewed']
}


df['payload_function'] = categorize_payload_functions(df, mapping)
```

```python
df['payload_outcome'] = df['payload_outcome'].str.replace(' Failure', ' failure', regex=False)

df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('ccess'), 'payload_outcome'] = 'Successful'
df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('deactivated'), 'payload_outcome'] = 'Successful'
df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('landed'), 'payload_outcome'] = 'Successful'
df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('docked'), 'payload_outcome'] = 'Successful'
df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('en route'), 'payload_outcome'] = 'Successful'
df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('operational'), 'payload_outcome'] = 'Operational'
df.loc[df['payload_outcome'].fillna('').str.lower().str.lower().str.contains('unclear'), 'payload_outcome'] = 'Unknown'
df.loc[df['payload_outcome'].fillna('').str.lower().str.lower().str.contains('destroyed'), 'payload_outcome'] = 'Failure'
df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('malfunction'), 'payload_outcome'] = 'Failure'
df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('fail'), 'payload_outcome'] = 'Failure'
df.loc[df['payload_outcome'].fillna('').str.lower().str.contains('intercepted'), 'payload_outcome'] = 'Failure'

possible_values = ['Successful', 'Operational', 'Failure', 'Unknown']
df.loc[~df['payload_outcome'].isin(possible_values), 'payload_outcome'] = 'Unknown'

df.groupby('payload_outcome')['flight_n'].nunique().sort_values(ascending=False).reset_index()
```

```python
df['payload_outcome_n'] = 0

df.loc[df['payload_outcome'].isin(['Successful', 'Operational']), 'payload_outcome_n'] = 2
df.loc[df['payload_outcome'] == 'Failure', 'payload_outcome_n'] = 1


df.groupby('payload_outcome_n')['flight_n'].nunique().sort_values(ascending=False).reset_index()
```

```python
rocket_columns = ['flight_n', 'launch_datetime', 'rocket', 'rocket_state', 'rocket_state_grouped', 'flight_number', 'launch_site', 'launch_site_state', 
                  'lsp', 'lsp_state']


payload_function = df[df['payload_function'].notnull()].groupby('flight_n')['payload_function'].agg(lambda x: pd.Series.mode(x)[0])
payload_orbit = df[df['payload_orbit'].notnull()].groupby('flight_n')['payload_orbit'].agg(lambda x: pd.Series.mode(x)[0])

boolen_fields = df.groupby('flight_n').agg({'is_luna': 'min', 'is_missile': 'max', 'payload_outcome_n': 'max'})

rockets_df = df[rocket_columns].drop_duplicates().set_index('flight_n')

rockets_df = pd.concat([rockets_df, payload_function, payload_orbit, boolen_fields], axis=1, join='outer').reset_index()

rockets_df['payload_outcome'] = 'Unknown'

rockets_df.loc[rockets_df['payload_outcome_n'] == 1, 'payload_outcome'] = 'Failure'
rockets_df.loc[rockets_df['payload_outcome_n'] == 2, 'payload_outcome'] = 'Successful'

rockets_df.drop('payload_outcome_n', axis=1, inplace=True)

rockets_df.head()
```
