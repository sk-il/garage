```python
import pandas as pd

directory = r'data\\'
```

```python
df = pd.read_excel(directory + 'Pulp Fiction.xlsx')
df.head()
```

```python
def count_occurrences(word, sentence):
    # Count occurrences of the word in the sentence, case insensitive
    return sentence.lower().count(word.lower())

# Adding a new column 'Fuck Count' to the DataFrame
df['Fuck Count 3'] = df['Subtitle'].apply(lambda x: count_occurrences('fuck', x))

# Displaying the DataFrame with the new column
df.head()
```

```python
df[df['Fuck Count'] != df['Fuck Count 3']]
```

```python
df['Fuck Count'].sum()
```

```python
df.drop(['Fuck Count 2', 'Fuck Count 3'], axis=1, inplace=True)
```

```python
subtitles = pd.read_excel(directory + 'subtitles.xlsx')
subtitles.head()
```

```python
subtitles = subtitles[['start_seconds', 'Subtitle', 'Fuck Count']].set_index('start_seconds')

df = df.join(subtitles, on='start_seconds', how='left', rsuffix=' 2')
```

```python
df[df['Fuck Count'] != df['Fuck Count 2']].to_excel(directory + 'fuck_check.xlsx', index=False)
```
