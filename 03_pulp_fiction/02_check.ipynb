```python
import pandas as pd

directory = r'data\\'
```

```python
df = pd.read_excel(directory + 'Pulp Fiction.xlsx')
df.head()
```

```python
def count_occurrences(word, sentence):
    # Count occurrences of the word in the sentence, case insensitive
    return sentence.lower().count(word.lower())

# Adding a new column 'Fuck Count' to the DataFrame
df['Fuck Count 3'] = df['Subtitle'].apply(lambda x: count_occurrences('fuck', x))

# Displaying the DataFrame with the new column
df.head()
```

```python
df[df['Fuck Count'] != df['Fuck Count 3']]
```

```python
df['Fuck Count'].sum()
```

```python
df.drop(['Fuck Count 2', 'Fuck Count 3', 'Check'], axis=1, inplace=True)

df['Word Count'] = df['Subtitle'].apply(lambda x: count_occurrences(' ', x)) + 1

df.head()
```

```python
df.info()
```

```python
df.to_excel(directory + 'pulp_fiction_final.xlsx', index=False, sheet_name='Subs')
```
