```python
import pandas as pd
```

```python
df = pd.read_csv(r'raw\boardgames_list.csv')
df.head()
```

```python
df.info() # 150057 
```

```python
link_duplicates = df.loc[df['link'].duplicated(), 'link']
df[df['link'].isin(link_duplicates)].sort_values('link').head(30)
```

```python
df.drop_duplicates(subset=['link'], keep='last', inplace=True)
len(df)
```

```python
df[df['rank'].isnull()].sort_values('num_voters', ascending=False).head()
```

```python
df[df['link'].str.contains('boardgameexpansion')].sort_values('rank').head()
```

```python
df = df[~df['link'].str.contains('boardgameexpansion')]
len(df)
```

```python
df[df['rank'].isnull()].sort_values('num_voters', ascending=False).head(30)
```

```python
df = df[~df['rank'].isnull()]
len(df)
```

```python
df.info()
```

```python
df['rank'] = df['rank'].astype(int)
df['year'] = df['year'].astype('Int64')
df['num_voters'] = df['num_voters'].astype(int)

df['id'] = df['link'].str.split('/', expand=True)[4].astype(int)
```

```python
df[df['price'].notnull() & ~df['price'].fillna('').str.contains('€')]
```

```python
df[df['price'].notnull() & ~df['price'].fillna('').str.contains('Amazon')]
```

```python
df['price_euro'] = df['price'].str.split(expand=True)[1].str.replace('.', '', regex=False)\
                              .str.replace(',', '.', regex=False).astype(float)
```

```python
df[df['price'].notnull() & df['price_euro'].isnull()]
```

```python
df.columns
```

```python
columns = ['id', 'title', 'rank', 'link', 'year', 'geek_rating', 'avg_rating', 'num_voters', 'description', 'price_euro']
df = df[columns]

df.head()
```
